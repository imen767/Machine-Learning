```python
# definir les colonnes numeriques et categoriels
numeric_features = df.select_dtypes(include=['int64','float64']).columns.tolist()
categorical_features = df.select_dtypes(include=['object','string','category']).columns.tolist()
print(f"On a {len(numeric_features)} features numériques : {numeric_features}")
print(f"\nOn a {len(categorical_features)} features catégorielles : {categorical_features}")
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

colonnes = ['Age', 'AvgDaysBetweenPurchases', 
            'SupportTicketsCount', 'SatisfactionScore']

for i, col in enumerate(colonnes):
    ax = axes[i//2, i%2]
    df[col].hist(bins=30, ax=ax, color='steelblue', edgecolor='white')
    ax.axvline(df[col].mean(), color='red', linestyle='--', label=f'Moyenne = {df[col].mean():.1f}')
    ax.axvline(df[col].median(), color='green', linestyle='--', label=f'Médiane = {df[col].median():.1f}')
    ax.set_title(col)
    ax.legend()

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
df['Recency'].hist(bins=30, color='steelblue', edgecolor='white')
plt.title('Distribution Recency')
plt.xlabel('Jours depuis dernier achat')
plt.ylabel('Nombre de clients')

plt.subplot(1, 2, 2)
df['Frequency'].hist(bins=30, color='coral', edgecolor='white')
plt.title('Distribution Frequency')
plt.xlabel('Nombre de commandes')
plt.ylabel('Nombre de clients')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 4))

# Comptage
churn_counts = df['Churn'].value_counts()
colors = ['steelblue', 'coral']

plt.subplot(1, 2, 1)
churn_counts.plot(kind='bar', color=colors, edgecolor='white')
plt.title('Répartition du Churn')
plt.xlabel('0 = Fidèle  |  1 = Parti')
plt.ylabel('Nombre de clients')
plt.xticks(rotation=0)

plt.subplot(1, 2, 2)
plt.pie(churn_counts, labels=['Fidèle', 'Parti'], colors=colors,
        autopct='%1.1f%%', startangle=90)
plt.title('% Churn')

plt.tight_layout()
plt.show()

print(f"Clients fidèles : {churn_counts[0]} ({churn_counts[0]/len(df)*100:.1f}%)")
print(f"Clients partis  : {churn_counts[1]} ({churn_counts[1]/len(df)*100:.1f}%)")
```
